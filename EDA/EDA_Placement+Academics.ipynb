```python
for i in range(0,len(df)): 
    for j in range(0,len(df['CourseName'][i])):
        df['CourseName'][i][j]=df['CourseName'][i][j].lower()
```

```python
df["ElectiveList"] = ""
df["ElectiveGrade"] = ""
```

```python
for i in range(0,len(df)): 
    el=[]
    grade=[]
    for j in range(0,len(df['CourseName'][i])):
        if(df['CourseName'][i][j] in electives):
            el.append(df['CourseName'][i][j])
            grade.append(df['CourseGrade'][i][j])
    df['ElectiveList'][i]=el
    df['ElectiveGrade'][i]=grade
```
