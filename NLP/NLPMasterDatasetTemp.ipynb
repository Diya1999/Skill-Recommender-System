```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv (r'MasterDataset_2016_2017_temp.csv') 
```

```python
df.drop(['Placed'],axis='columns', inplace=True)
```

```python
# 10th percentage and 12 percentage 
# convert everything to percent 10thPercentage, rename 10th marks
# null values -> mean
df['10thMarks'] = 0
df['10thMarks'] = df['10thPercentage'].map(lambda x: float(x)*9.5 if float(x) <=10 else float(x) )
df['10thMarks'].fillna((df['10thMarks'].mean()), inplace=True)
df['12thPercentage'].fillna((df['12thPercentage'].mean()), inplace=True)
```

```python
df['Awards'].fillna((''),inplace = True)
df['Awards'] = df['Awards'].str.lower()
df['Languages'].fillna((''),inplace = True)
df['ProgLanguages'].fillna((''),inplace = True)
```

```python
# awards, using keywords to extract

from nltk import ngrams
from sklearn.feature_extraction import text
stop = text.ENGLISH_STOP_WORDS

df["Awards"].apply(lambda words: ' '.join(word.lower() for word in words.split() if word not in stop))

vals = [y for x in df['Awards'] for y in x.split()]

n = [2,3]
a = pd.Series([y for x in n for y in ngrams(vals, x)]).value_counts()
a[:10]
```
