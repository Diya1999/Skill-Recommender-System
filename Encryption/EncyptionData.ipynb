```python
import pandas as pd 
from itsdangerous import URLSafeSerializer
```

```python
data = pd.read_csv("OfferdStudents.csv") 
```

```python
def encrypt(text,s):
    result = ""
    for i in range(len(text)):
        char = text[i]
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
    return result
s = 2345
```

```python
l1=[]
l2=[]
for i in data['USN']:
    l1.append(encrypt(i,s))
for i in data['Name (As per the Academic records)']:
    l2.append(encrypt(i,s))
```

```python
data['USN']=l1
data['Name (As per the Academic records)']=l2
```

```python
data.to_csv(r'OfferedStudentsEncypted.csv')
```
