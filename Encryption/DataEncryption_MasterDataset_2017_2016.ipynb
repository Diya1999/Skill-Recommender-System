```python
def encrypt(text,s):
    result = ""
    for i in range(len(text)):
        char = text[i]
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
    return result
s = 23455
```

```python
l1=[]
for i in data['USN']:
    l1.append(encrypt(i,s))
data['USN']=l1
```
