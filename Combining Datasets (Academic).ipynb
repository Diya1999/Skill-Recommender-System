```python
usn_list = final_df.USN.unique()
print(usn_list.size)
```

```python
final_df_cs = final_df[final_df['USN'].isin(usn_list_cs)]
```

```python
final_df_cs.shape
final_df_cs.USN.unique().size
```

```python
final_df_cs['Course'] = final_df_cs.apply(lambda x: {"Code":x['SubjectCode'],"Name":x['SubjectName'],"Credits":x['Credits']},axis=1)   
```

```python
final_df_cs['CourseGrade'] = final_df_cs.apply(lambda x: {str(x['Course']):x['Grade']},axis=1)   
```

```python
final_df_cs
```

```python
grouped = final_df_cs.groupby(['USN','elective'])
grouped_df = grouped.aggregate(lambda x: list(x))
```

```python
grouped_df_reset = grouped_df.reset_index()
grouped_df_reset
```

```python
grades_total = grouped_df_reset.groupby('USN').apply(lambda x: dict(zip(x['elective'],x['CourseGrade'])))
```

```python
type(grades_total)
```

```python
grades_total.to_frame().reset_index()
grades_total_final = grades_total.to_frame().reset_index().rename(columns = {0: 'Courses'}, inplace = False)
```

```python
grades_total_final.to_csv(r'CoursesCS.csv', index = False)
```
