```python
import numpy, csv
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
#Target: CGPA, earned credits
#Skills: Company name, tier, CTC
df = pd.read_csv('C:\\Users\\diyas\\Documents\\Recommendation_system\\Placement+Academics1516_UpdatedTier.csv',skipinitialspace=True) 
```

```python
#removing unnessary columns
df=df.drop(['Mobile','Name','Email','Stipend(K)','EmploymentType','TotalCredits','StudentId','TierLevel','Number of Offers'], axis = 1) 
```

```python
'''for i in ['Number of Offers']: 
    df[i] = df[i].replace(np.nan,"Number of Offer Unavailable") 
    df[i] = df[i].replace(" ","Number of Offer Unavailable") '''
    
for i in ['CTC']: 
    df[i] = df[i].replace(np.nan,"CTC Unavailable") 
    df[i] = df[i].replace(" ","CTC Unavailable")
                          
'''for i in ['TierLevel']: 
    df[i] = df[i].replace(np.nan,"TierLevel Unavailable") 
    df[i] = df[i].replace(" ","TierLevel Unavailable")'''
    
for i in ['CompanyName']: 
    df[i] = df[i].replace(np.nan,"CompanyName Unavailable") 
    df[i] = df[i].replace(" ","CompanyName Unavailable")
```

```python
records = []
for i in range(0, 1055):
    records.append([str(df.values[i,j]) for j in range(0, 6)])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```
