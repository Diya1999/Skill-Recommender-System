```python
#academic information of computer science students from batch of 2015 and 2016
academic = pd.read_csv(r"C:\Users\rames\OneDrive\Desktop\namma\FinalYear\AcademicCombined_15_16_CS.csv")
```

```python
placement = pd.read_csv(r"C:\Users\rames\OneDrive\Desktop\namma\FinalYear\PlacementCombined15_16.csv")
```

```python
print(academic.shape, placement.shape)
```

```python
placement['USN'] = placement['USN'].str.upper() 
academic['USN'] = academic['USN'].str.upper() 
```

```python
#combining 2015 and 2016 data for CSE branch
col_placement =[]
col_academic=[]
for col in placement.columns: 
    col_placement.append(col)

for col in academic.columns:
    col_academic.append(col)
print(col_placement,"\n",col_academic)
```

```python
acadPlac = academic
```

```python
placement['Number of offers'] =placement['USN'].map(placement['USN'].value_counts())
acadPlac['Number of offers'] = acadPlac['USN'].map(placement['USN'].value_counts())
```

```python
acadPlacNull = acadPlac[acadPlac['Number of offers'].isnull()]
acadPlacNotNull = acadPlac[acadPlac['Number of offers'].notnull()]
acadPlacNotNullRep = acadPlacNotNull.reindex(acadPlacNotNull.index.repeat(acadPlacNotNull['Number of offers']))
acadPlacRep = pd.concat([acadPlacNotNullRep, acadPlacNull])
acadPlacRep = acadPlacRep.sort_values(by='USN')
acadPlacRep.shape
```

```python
a = list(acadPlacRep.USN.unique())
b = list(placement.USN.unique())
list(set(b) - set(a))
```

```python
placementA = placement[~placement.USN.str.contains("01FB17")]
a = list(acadPlacRep.USN.unique())
b = list(placementA.USN.unique())
list(set(b) - set(a))
```

```python
acadPlacFinal = pd.merge(acadPlac, placementA, on='USN', how='outer').sort_values(by='USN')
```

```python
acadPlacFinal.drop('Number of offers_x',axis='columns', inplace=True)
acadPlacFinal.rename(columns = {'Number of offers_y':'Number of Offers'}, inplace = True)
```
