```python
import pandas as pd
import numpy as np
```

```python
grade_df = pd.read_csv (r'C:\Users\diyas\Downloads\Grades_17_Batch_Students_Grades_AllSubjects.csv') 
```

```python
grade_df.columns = grade_df.columns.str.replace(' ', '')
grade_df.columns = grade_df.columns.str.replace('EarnedCredit', 'EarnedCreditSubject')
```

```python
grade_df['elective'] = 'Core Courses'
```

```python
elective_list=['UE17CS331','UE17CS332','UE17CS333','UE17CS334','UE17CS335','UE17CS336','UE17CS337','UE17CS338','UE17CS341','UE17CS342','UE17CS343','UE17CS344','UE17CS345','UE17CS346','UE17CS411','UE17CS412','UE17CS413','UE17CS414','UE17CS421','UE17CS422','UE17CS423','UE17CS424','UE17CS347','UE17CS348','UE17CS311','UE17CS312','UE17CS313','UE17CS314','UE17CS315','UE17CS321','UE17CS322','UE17CS323','UE17CS324','UE17CS325']
```

```python
grade_df.loc[grade_df['SubjectCode'].isin(elective_list), 'elective'] = 'Y'
```

```python
grade_df
```

```python
grade_rev_df =  grade_df[grade_df.columns.difference(['EarnedCreditSubject', 'GPS'])]
```

```python
grade_rev_df.shape
```

```python
grade_df.shape
```

```python
grade_rev_df.head()
```

```python
grade_overall_rev_df=grade_rev_df
rem_duplicate = grade_overall_rev_df.drop_duplicates(subset=['USN', 'SubjectCode'], keep='first')
```

```python
rem_duplicate1 = grade_overall_rev_df.drop_duplicates(subset=['USN', 'SubjectCode','Grade'], keep='first')
```

```python
rem_duplicate1[~rem_duplicate1.apply(tuple,1).isin(rem_duplicate.apply(tuple,1))]
```

```python
final_df = rem_duplicate1
```

```python
final_df
```

```python
usn_list = final_df.USN.unique()
print(usn_list.size)
```

```python
final_df['Course'] = final_df.apply(lambda x: {"Code":x['SubjectCode'],"Name":x['SubjectName'],"Credits":x['Credits']},axis=1)   
```

```python
final_df['CourseGrade'] = final_df.apply(lambda x: {str(x['Course']):x['Grade']},axis=1)   
```

```python
final_df
```

```python
grouped = final_df.groupby(['USN','elective'])
grouped_df = grouped.aggregate(lambda x: list(x))
```

```python
grouped_df_reset = grouped_df.reset_index()
grouped_df_reset
```

```python
grades_total = grouped_df_reset.groupby('USN').apply(lambda x: dict(zip(x['elective'],x['CourseGrade'])))
```

```python
grades_total.to_frame().reset_index()
grades_total_final = grades_total.to_frame().reset_index().rename(columns = {0: 'Courses'}, inplace = False)
```

```python
grades_total_final
```

```python
grades_total_final.to_csv(r'2017Courses.csv', index = False)
```
