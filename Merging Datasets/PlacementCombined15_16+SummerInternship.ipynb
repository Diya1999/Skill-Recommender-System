```python
summer_intern_16.rename(columns = {'Name (As per the Academic records)':'Name'}, inplace = True)
summer_intern_16.rename(columns = {'Summer Internship Offered @':'CompanyName'},inplace = True)
summer_intern_16.rename(columns = {'Tier':'TierLevel'}, inplace = True)
```

```python
#adding a new column for EmploymentType
summer_intern_16['EmploymentType'] = 'Summer Internship'
```

```python
#adding stipend column for summer internships, extracting values in brackets
a = summer_intern_16['CompanyName'].str.split('(', 1).to_list()
stipend =[]
company = []
for i in a:
    if(i[1]=='CCBD)'):
        stipend.append(0)
    else:
        stipend.append(i[1].split('K)')[0])
    company.append(i[0])
stipendDf = pd.DataFrame(stipend,columns=["Stipend(K)"])
companyName = pd.DataFrame(company,columns = ["companyName"])
#stipendDf
```

```python
summer_intern_16A = summer_intern_16.reset_index(drop=True)
stipendDfA = stipendDf.reset_index(drop=True)
companyA = companyName.reset_index(drop=True)
```

```python
summerIntern16 = summer_intern_16A.join(stipendDfA).join(companyA)
```

```python
summerIntern16.drop('CompanyName',axis='columns', inplace=True)
summerIntern16.rename(columns = {'companyName':'CompanyName'}, inplace = True)
```
